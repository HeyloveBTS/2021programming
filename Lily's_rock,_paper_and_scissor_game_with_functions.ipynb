##RPS
1.computer choose randomly between rock, paper and scissor

a. rock<paper, paper<scissor, scissor<rock

2.user choose between rock, paper and scissor

3.if user < computer, user lose if user == computer, print play again if user > computer, user win


for step a:


* player 1 chooses scissors and player 2 chooses rock, player 2 wins
* player 1 chooses paper and player 2 chooses rock, player 1 wins 
* player 1 chooses paper  and player 2 chooses scissors, player 2 wins 
* player 2 chooses scissors and player 1 chooses rock, player 1 wins 
* player 2 chooses paper and player 1 chooses rock, player 2 wins 
* player 2 chooses paper  and player 1 chooses scissors, player 1 wins 
* Both players choose the same object, play again


```python
import random
rock = "rock"
paper = "paper"
scissor = "scissor"
RPS = [rock,paper,scissor]
compMove = random.choice(RPS)


def user_Move():
  userMove = input("rock, paper or scissor: ")
  if str(userMove) == "rock":
    print("you selected rock")
    userMove = rock
  elif str(userMove) == "paper":
    userMove = paper
    print("you selected paper")
  else:
    print("you selected scissor")
    userMove = scissor

def comp_move():
  print("the computer selected", compMove)

def compare():
  i = True
  if userMove == compMove:
    print("It's a tie!")
    i = False
  elif userMove == rock:
    if compMove == scissor:
      print("You won")
      i = False
    else:
      print("You lose")
      i = False
  elif userMove == scissor:
    if compMove == paper:
      print("You won")
      i = False
    else:
      print("You lose")
      i = False
  elif userMove == paper:
    if compMove == rock:
      print("You won")
      i = False
    else:
      print("You lose")
      i = False

user_Move()
comp_move()
compare()
```
